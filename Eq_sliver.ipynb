# <B>Earthquake Events - Sliver layer

```python
from datetime import date, timedelta
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType
```

#### <b>Reading the bronze layer file for processing

```python
# Reading the file which we created in bronze layer

df=spark.read.option("multiline","true").json(f"Files/{start_date}_earthquake_date.json")
```

```python
display(df)
```

#### <B>Extracting the data from json dump

```python
### Extracting the data from the json dump
df = df.select(
    'id',
    col('geometry.coordinates').getItem(0).alias('longituede'),
    col('geometry.coordinates').getItem(1).alias('latitude'),
    col('geometry.coordinates').getItem(1).alias('elevation'),
    col('properties.title').alias('title'),
    col('properties.felt').alias('earthquak_felt'),
    col('properties.alert').alias('earthquake_alert'),
    col('properties.sig').alias('strenth_of_earthquake'),
    col('properties.updated').alias('update_time'),
    col('properties.time').alias('time'),
    col('properties.mag').alias('magnitude'),
    col('properties.tsunami').alias('tsunami'),
    col('properties.magType').alias('magnitude_type'),
)
```

#### <B>Displaying the dataframe after processing


```python
display(df)
```

#### <b>Working on the "Update Time" which is millisecond converting it to Data and time stamp

```python
df=df.\
    withColumn('time',col('time')/1000).\
    withColumn('update_time',col('update_time')/1000).\
    withColumn('time',col('time').cast(TimestampType())).\
    withColumn('update_time',col('update_time').cast(TimestampType()))
```

```python
display(df)
```

#### <B>Writing dataframe in to table

```python
## Writing the dataframe as table in Tables folder in lakehouse
df.write.mode('append').saveAsTable('earthquake_sliver')
```
